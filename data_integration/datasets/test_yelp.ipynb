```python
import json
import pandas as pd
from tqdm import tqdm
```

```python
business = open("archive/yelp_academic_dataset_business.json")
data = []
for line in tqdm(business):
    data.append(json.loads(line))
business_df = pd.DataFrame(data)
business.close()
business_df
```

```python
for row in business_df['attributes'].iloc[:10]:
    print(row)
    try:
        for key, value in row.items():
            print(f'{key}: {value}', end='; ')
        print()
    except:
        continue
```

```python
for row in business_df['hours'].iloc[:10]:
    print(row)
    try:
        for key, value in row.items():
            print(f'{key}: {value}', end='; ')
        print()
    except:
        continue
```

```python
for index, row in business_df.iloc[:10].iterrows():
    print(f"business: {row['business_id']}; categories:", end=' ')
    for category in row['categories'].split(', '):
        print(f'{category}', end=',')
    print()
```

```python
checkin = open("archive/yelp_academic_dataset_checkin.json")
data = []
for line in tqdm(checkin):
    data.append(json.loads(line))
checkin_df = pd.DataFrame(data)
checkin.close()
checkin_df
```

```python
review = open("archive/yelp_academic_dataset_review.json")
data = []
for index, line in tqdm(enumerate(review)):
    data.append(json.loads(line))
    if index > 100000:
        break
review_df = pd.DataFrame(data)
review.close()
review_df
```

```python
tip = open("archive/yelp_academic_dataset_tip.json")
data = []
for index, line in enumerate(tqdm(tip)):
    data.append(json.loads(line))
    if index > 100000:
        break
tip_df = pd.DataFrame(data)
tip.close()
tip_df
```

```python
user = open("archive/yelp_academic_dataset_user.json")
data = []
for index, line in tqdm(enumerate(user)):
    data.append(json.loads(line))
    if index > 100000:
        break
user_df = pd.DataFrame(data)
user.close()
user_df
```

```python
for index, row in user_df.iloc[:10].iterrows():
    print(f"user: {row['user_id']}; friends:", end=' ')
    for friend in row['friends'].split(', '):
        print(f'{friend}', end=',')
    print()
```
